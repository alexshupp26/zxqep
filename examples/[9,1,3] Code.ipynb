```python
import pyzx as zx

g = zx.Graph()

# ==========================================
# 0. INPUT BOUNDARIES (Data + 8 Ancillas)
# ==========================================
v0_0 = g.add_vertex(0, 0, 0)
v1_0 = g.add_vertex(2, 1, 0)
v2_0 = g.add_vertex(2, 2, 0)
v3_0 = g.add_vertex(2, 3, 0)
v4_0 = g.add_vertex(2, 4, 0)
v5_0 = g.add_vertex(2, 5, 0)
v6_0 = g.add_vertex(2, 6, 0)
v7_0 = g.add_vertex(2, 7, 0)
v8_0 = g.add_vertex(2, 8, 0)

v9_0 = g.add_vertex(2, 9, 0)
v10_0 = g.add_vertex(2, 10, 0)
v11_0 = g.add_vertex(2, 11, 0)
v12_0 = g.add_vertex(2, 12, 0)
v13_0 = g.add_vertex(2, 13, 0)
v14_0 = g.add_vertex(2, 14, 0)
v15_0 = g.add_vertex(1, 15, 0)
v16_0 = g.add_vertex(1, 16, 0)

# ==========================================
# 1. BIT-FLIP (ZZ) SYNDROME EXTRACTION
# ==========================================
# Row 1: First half of ZZ checks for blocks 1, 2, and 3
v0_1 = g.add_vertex(1, 0, 1); v9_1 = g.add_vertex(2, 9, 1); g.add_edge((v0_1, v9_1))
v3_1 = g.add_vertex(1, 3, 1); v11_1 = g.add_vertex(2, 11, 1); g.add_edge((v3_1, v11_1))
v6_1 = g.add_vertex(1, 6, 1); v13_1 = g.add_vertex(2, 13, 1); g.add_edge((v6_1, v13_1))

# Row 2: Second half of ZZ checks for blocks 1, 2, and 3
v1_2 = g.add_vertex(1, 1, 2); v9_2 = g.add_vertex(2, 9, 2); g.add_edge((v1_2, v9_2))
v4_2 = g.add_vertex(1, 4, 2); v11_2 = g.add_vertex(2, 11, 2); g.add_edge((v4_2, v11_2))
v7_2 = g.add_vertex(1, 7, 2); v13_2 = g.add_vertex(2, 13, 2); g.add_edge((v7_2, v13_2))

# Row 3: First half of the adjacent ZZ checks
v1_3 = g.add_vertex(1, 1, 3); v10_3 = g.add_vertex(2, 10, 3); g.add_edge((v1_3, v10_3))
v4_3 = g.add_vertex(1, 4, 3); v12_3 = g.add_vertex(2, 12, 3); g.add_edge((v4_3, v12_3))
v7_3 = g.add_vertex(1, 7, 3); v14_3 = g.add_vertex(2, 14, 3); g.add_edge((v7_3, v14_3))

# Row 4: Second half of the adjacent ZZ checks
v2_4 = g.add_vertex(1, 2, 4); v10_4 = g.add_vertex(2, 10, 4); g.add_edge((v2_4, v10_4))
v5_4 = g.add_vertex(1, 5, 4); v12_4 = g.add_vertex(2, 12, 4); g.add_edge((v5_4, v12_4))
v8_4 = g.add_vertex(1, 8, 4); v14_4 = g.add_vertex(2, 14, 4); g.add_edge((v8_4, v14_4))

# ==========================================
# 2. PHASE-FLIP (XXXXXX) SYNDROME EXTRACTION
# ==========================================
# Row 5: Prepare X-ancillas in |+> state
v15_5 = g.add_vertex(3, 15, 5)
v16_5 = g.add_vertex(3, 16, 5)

# Row 6-11: Waterfall CNOTs from Ancillas to Data
v0_6 = g.add_vertex(2, 0, 6); v15_6 = g.add_vertex(1, 15, 6); g.add_edge((v0_6, v15_6))
v3_6 = g.add_vertex(2, 3, 6); v16_6 = g.add_vertex(1, 16, 6); g.add_edge((v3_6, v16_6))

v1_7 = g.add_vertex(2, 1, 7); v15_7 = g.add_vertex(1, 15, 7); g.add_edge((v1_7, v15_7))
v4_7 = g.add_vertex(2, 4, 7); v16_7 = g.add_vertex(1, 16, 7); g.add_edge((v4_7, v16_7))

v2_8 = g.add_vertex(2, 2, 8); v15_8 = g.add_vertex(1, 15, 8); g.add_edge((v2_8, v15_8))
v5_8 = g.add_vertex(2, 5, 8); v16_8 = g.add_vertex(1, 16, 8); g.add_edge((v5_8, v16_8))

v3_9 = g.add_vertex(2, 3, 9); v15_9 = g.add_vertex(1, 15, 9); g.add_edge((v3_9, v15_9))
v6_9 = g.add_vertex(2, 6, 9); v16_9 = g.add_vertex(1, 16, 9); g.add_edge((v6_9, v16_9))

v4_10 = g.add_vertex(2, 4, 10); v15_10 = g.add_vertex(1, 15, 10); g.add_edge((v4_10, v15_10))
v7_10 = g.add_vertex(2, 7, 10); v16_10 = g.add_vertex(1, 16, 10); g.add_edge((v7_10, v16_10))

v5_11 = g.add_vertex(2, 5, 11); v15_11 = g.add_vertex(1, 15, 11); g.add_edge((v5_11, v15_11))
v8_11 = g.add_vertex(2, 8, 11); v16_11 = g.add_vertex(1, 16, 11); g.add_edge((v8_11, v16_11))

# Row 12: Measure X-ancillas in X-basis
v15_12 = g.add_vertex(3, 15, 12)
v16_12 = g.add_vertex(3, 16, 12)

# ==========================================
# 3. OUTPUT BOUNDARIES
# ==========================================
v0_13 = g.add_vertex(0, 0, 13)
v1_13 = g.add_vertex(0, 1, 13)
v2_13 = g.add_vertex(0, 2, 13)
v3_13 = g.add_vertex(0, 3, 13)
v4_13 = g.add_vertex(0, 4, 13)
v5_13 = g.add_vertex(0, 5, 13)
v6_13 = g.add_vertex(0, 6, 13)
v7_13 = g.add_vertex(0, 7, 13)
v8_13 = g.add_vertex(0, 8, 13)

v9_13 = g.add_vertex(2, 9, 13)
v10_13 = g.add_vertex(2, 10, 13)
v11_13 = g.add_vertex(2, 11, 13)
v12_13 = g.add_vertex(2, 12, 13)
v13_13 = g.add_vertex(2, 13, 13)
v14_13 = g.add_vertex(2, 14, 13)
v15_13 = g.add_vertex(1, 15, 13)
v16_13 = g.add_vertex(1, 16, 13)

# ==========================================
# 4. HORIZONTAL CONNECTIONS
# ==========================================
# Data Qubits
g.add_edges([(v0_0, v0_1), (v0_1, v0_6), (v0_6, v0_13)])
g.add_edges([(v1_0, v1_2), (v1_2, v1_3), (v1_3, v1_7), (v1_7, v1_13)])
g.add_edges([(v2_0, v2_4), (v2_4, v2_8), (v2_8, v2_13)])
g.add_edges([(v3_0, v3_1), (v3_1, v3_6), (v3_6, v3_9), (v3_9, v3_13)])
g.add_edges([(v4_0, v4_2), (v4_2, v4_3), (v4_3, v4_7), (v4_7, v4_10), (v4_10, v4_13)])
g.add_edges([(v5_0, v5_4), (v5_4, v5_8), (v5_8, v5_11), (v5_11, v5_13)])
g.add_edges([(v6_0, v6_1), (v6_1, v6_9), (v6_9, v6_13)])
g.add_edges([(v7_0, v7_2), (v7_2, v7_3), (v7_3, v7_10), (v7_10, v7_13)])
g.add_edges([(v8_0, v8_4), (v8_4, v8_11), (v8_11, v8_13)])

# ZZ Ancillas
g.add_edges([(v9_0, v9_1), (v9_1, v9_2), (v9_2, v9_13)])
g.add_edges([(v10_0, v10_3), (v10_3, v10_4), (v10_4, v10_13)])
g.add_edges([(v11_0, v11_1), (v11_1, v11_2), (v11_2, v11_13)])
g.add_edges([(v12_0, v12_3), (v12_3, v12_4), (v12_4, v12_13)])
g.add_edges([(v13_0, v13_1), (v13_1, v13_2), (v13_2, v13_13)])
g.add_edges([(v14_0, v14_3), (v14_3, v14_4), (v14_4, v14_13)])

# XXXXXX Ancillas
g.add_edges([(v15_0, v15_5), (v15_5, v15_6), (v15_6, v15_7), (v15_7, v15_8), (v15_8, v15_9), (v15_9, v15_10), (v15_10, v15_11), (v15_11, v15_12), (v15_12, v15_13)])
g.add_edges([(v16_0, v16_5), (v16_5, v16_6), (v16_6, v16_7), (v16_7, v16_8), (v16_8, v16_9), (v16_9, v16_10), (v16_10, v16_11), (v16_11, v16_12), (v16_12, v16_13)])
```
