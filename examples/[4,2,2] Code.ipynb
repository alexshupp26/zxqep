# [4,2,2] Code

This is an example involving the QEC code that encodes 2 logical qubits into 4 physical qubits

```python
import pyzx as zx
import zxqep as ep 
```

## Create your QEC Code

```python
### Create PyZX Graph 
g = zx.Graph()

# Add Inputs
v0_0 = g.add_vertex(1, 0, 0)
v1_0 = g.add_vertex(2, 1, 0)
v2_0 = g.add_vertex(2, 2, 0)
v3_0 = g.add_vertex(2, 3, 0)

## State Preparation Circuit
# CNOT12
v0_1 = g.add_vertex(1, 0, 1)
v1_1 = g.add_vertex(2, 1, 1)
g.add_edge((v0_1, v1_1))

# CNOT13
v0_2 = g.add_vertex(1, 0, 2)
v2_2 = g.add_vertex(2, 2, 2)
g.add_edge((v0_2, v2_2))

# CNOT14
v0_3 = g.add_vertex(1, 0, 3)
v3_3 = g.add_vertex(2, 3, 3)
g.add_edge((v0_3, v3_3))


# Add Ancillas
v4_0 = g.add_vertex(1, 4, 0) 
v5_0 = g.add_vertex(2, 5, 0)

# Add XXXX Stabilizer check
v4_4 = g.add_vertex(1, 4, 4)
v0_4 = g.add_vertex(2, 0, 4) 
g.add_edge((v4_4, v0_4))

v4_5 = g.add_vertex(1, 4, 5) 
v1_5 = g.add_vertex(2, 1, 5)
g.add_edge((v4_5, v1_5))

v4_6 = g.add_vertex(1, 4, 6) 
v2_6 = g.add_vertex(2, 2, 6)
g.add_edge((v4_6, v2_6))

v4_7 = g.add_vertex(1, 4, 7) 
v3_7 = g.add_vertex(2, 3, 7)
g.add_edge((v4_7, v3_7))

# Add ZZZZ Stabilizer Check 
v0_8 = g.add_vertex(1, 0, 8) 
v5_8 = g.add_vertex(2, 5, 8) 
g.add_edge((v0_8, v5_8))

v1_9 = g.add_vertex(1, 1, 9)
v5_9 = g.add_vertex(2, 5, 9)
g.add_edge((v1_9, v5_9))

v2_10 = g.add_vertex(1, 2, 10)
v5_10 = g.add_vertex(2, 5, 10)
g.add_edge((v2_10, v5_10))

v3_11 = g.add_vertex(1, 3, 11)
v5_11 = g.add_vertex(2, 5, 11)
g.add_edge((v3_11, v5_11))

# Add Outputs
v0_12 = g.add_vertex(0, 0, 12)
v1_12 = g.add_vertex(0, 1, 12)
v2_12 = g.add_vertex(0, 2, 12)
v3_12 = g.add_vertex(0, 3, 12)

# Add ancilla measurements 
v4_12 = g.add_vertex(1, 4, 12)
v5_12 = g.add_vertex(2, 5, 12)

# Add Horizontal connections
g.add_edges([
    (v0_0, v0_1), (v0_1, v0_2), (v0_2, v0_3), (v0_3, v0_4), (v0_4, v0_8), (v0_8, v0_12)
])
g.add_edges([
    (v1_0, v1_1), (v1_1, v1_5), (v1_5, v1_9), (v1_9, v1_12)
])
g.add_edges([
    (v2_0, v2_2), (v2_2, v2_6), (v2_6, v2_10), (v2_10, v2_12)
])
g.add_edges([
    (v3_0, v3_3), (v3_3, v3_7), (v3_7, v3_11), (v3_11, v3_12)
])
g.add_edges([
    (v4_0, v4_4), (v4_4, v4_5), (v4_5, v4_6), (v4_6, v4_7), (v4_7, v4_12)
])
g.add_edges([
    (v5_0, v5_8), (v5_8, v5_9), (v5_9, v5_10), (v5_10, v5_11), (v5_11, v5_12)
])
```

```python
zx.draw(g)
```

Using PyZX's built-in simplification rules, this simplifies to a four qubit bell state: $|\psi \rangle = \frac{1}{2}(|0000\rangle + |1111\rangle)$

```python
g_vis = g.clone()
zx.full_reduce(g_vis)
zx.draw(g_vis)
```

## Inject, propagate, and visualize your errors in your QEC code

#### Bit Flip Error (Pauli-X) on the first qubit AFTER the state preparation

```python
# Create your ZXQEPGraph object
g_error = ep.ZXQEPGraph(g)

# Inject your starting error
g_error.inject_error(row=0, col = 4, err_type='X', fraction_val=1)

# View the circuit BEFORE propagation
print("--- Initial State (t=0) ---")
g_error.draw()

# Propagate the errror 
g_error.propagate_to_end()

# View the circuit AFTER propagation
print("--- Final State (t=end) ---")
g_error.draw()
```
