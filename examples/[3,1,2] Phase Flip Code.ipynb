# [3,1,2] Phase Flip Code

This is a simple example involving the 3-qubit repition code that protects against a single phase flip

```python
import pyzx as zx
import zxqep as ep
```

## Create your QEC Code

```python
### Create PyZX Graph
g = zx.Graph()


# Add Inputs
v00 = g.add_vertex(0,0,0)
v10 = g.add_vertex(2,1,0)
v20 = g.add_vertex(2,2,0)

## State Preparation Circuit

# Add Ancillas
v30 = g.add_vertex(2,3,0)
v40 = g.add_vertex(2,4,0)

# Add CNOT12
v01 = g.add_vertex(1,0,1)
v11 = g.add_vertex(2,1,1)
g.add_edge((v01,v11))

# Add CNOT13
v02 = g.add_vertex(1,0,2)
v22 = g.add_vertex(2,2,2)
g.add_edge((v02,v22))

# Add Hadamards 
v03 = g.add_vertex(3,0,3)
v13 = g.add_vertex(3,1,3)
v23 = g.add_vertex(3,2,3)

## Stabilizer Measurement Circuit

# Add Hadamards  
v04 = g.add_vertex(3,0,4)
v14 = g.add_vertex(3,1,4)
v24 = g.add_vertex(3,2,4)

# Add first parity check (do qubits 1 and 2 have the same phase)
v05 = g.add_vertex(1,0,5)
v35 = g.add_vertex(2,3,5)
g.add_edge((v05,v35))

v16 = g.add_vertex(1,1,6)
v36 = g.add_vertex(2,3,6)
g.add_edge((v16,v36))

# Add second parity check (do qubits 2 and 3 have the same phase)
v17 = g.add_vertex(1,1,7)
v47 = g.add_vertex(2,4,7)
g.add_edge((v17,v47))

v28 = g.add_vertex(1,2,8)
v48 = g.add_vertex(2,4,8)
g.add_edge((v28,v48))

# Add Hadamards 
v09 = g.add_vertex(3,0,9)
v19 = g.add_vertex(3,1,9)
v29 = g.add_vertex(3,2,9)

# Add outputs
v0_10 = g.add_vertex(0,0,10)
v1_10 = g.add_vertex(0,1,10)
v2_10 = g.add_vertex(0,2,10)

# Add ancilla measurements
v3_10 = g.add_vertex(2,3,10)
v4_10 = g.add_vertex(2,4,10)

# Connecting horizontal edges wire by wire
g.add_edges([(v00,v01), (v01,v02), (v02,v03), (v03,v04), (v04,v05), (v05,v09), (v09,v0_10)])
g.add_edges([(v10,v11), (v11,v13), (v13,v14), (v14,v16), (v16,v17), (v17,v19), (v19,v1_10)])
g.add_edges([(v20,v22), (v22,v23), (v23,v24), (v24,v28), (v28,v29), (v29,v2_10)])
g.add_edges([(v30,v35), (v35,v36), (v36,v3_10)])
g.add_edges([(v40,v47), (v47,v48), (v48,v4_10)])

zx.draw(g)
```

Using PyZX's built-in simplification rules, one can see that this simplifies to some arbitrary input $ | \psi \rangle = \alpha |0\rangle + \beta |1\rangle $ encoded into 3 physical qubits such that $|\psi \rangle = \alpha |+++\rangle + \beta |---\rangle$

```python
g_vis = g.clone()
zx.hsimplify.from_hypergraph_form(g_vis)
zx.full_reduce(g_vis)
zx.draw(g_vis)
```

## Inject, propagate, and visualize your errors in your QEC code

#### Phase Flip Error (Pauli-Z) on the first qubit AFTER the state preparation

```python
# Create your ZXQEPGraph object
g_error = ep.ZXQEPGraph(g)

# Inject your error
g_error.inject_error(row=0, col = 4, err_type='Z', fraction_val=1)

# View the circuit BEFORE propagation
print("--- Initial State (t=0) ---")
g_error.draw()

# Propagate the errror 
g_error.propagate_to_end()

# View the circuit AFTER propagation
print("--- Final State (t=end) ---")
g_error.draw()
```

How do these error(s) affect the final state?

```python
g_pyzx = g_error.to_pyzx_graph()
zx.hsimplify.from_hypergraph_form(g_pyzx)
zx.full_reduce(g_pyzx)
zx.draw(g_pyzx)
```
